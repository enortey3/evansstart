```python
import numpy as np

def f(x):
    return np.cos(x) - x / 2  # Example function

def fp(x):
    return -np.sin(x) - 1 / 2  # Derivative of the example function

def newton_method(f, fp, x0, niterates):
    x_current = x0
    iterates = [x_current]

    for i in range(niterates):
        x_next = x_current - f(x_current) / fp(x_current)  # Newton's update
        iterates.append(x_next)
        x_current = x_next  # Update current value for next iteration

    return iterates

# Call the newton_method function
iterates = newton_method(f, fp, x0=-2.22, niterates=10)
print(iterates)
```

```python


import numpy as np
import matplotlib.pyplot as plt

# Define the function for Secant method
def secant_method(f, x0, x1, xdomain, niterates):
    xmin, xmax = xdomain
    
    # Initialize vector of x-values for plot
    npts = 500
    x = np.linspace(xmin, xmax, npts)
    ymin = np.min(f(x))
    ymax = np.max(f(x))
    dy = (ymax - ymin) / 10
    
    # Plot the curve in blue
    plt.plot(x, f(x), 'b', label='f(x)')
    
    # Plot the x-axis and y-axis
    plt.plot([xmin, xmax], [0, 0], 'k')  # x-axis
    plt.plot([0, 0], [ymin - 2 * dy, ymax + 2 * dy], 'k')  # y-axis
    
    # Plot initial points
    plt.plot([x0, x0], [0, f(x0)], 'r--', label='initial value at x0')
    plt.plot([x1, x1], [0, f(x1)], 'r--', label='initial value at x1')
    
    # Compute the next approximation using the secant formula
    x_next = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
    
    # Plot the secant line segment
    plt.plot([x0, x1], [f(x0), f(x1)], 'r-')
    plt.plot([x1, x_next], [f(x1), 0], 'r-')
    plt.plot([x_next, x_next], [0, f(x_next)], 'r--')

    # Show grid and set axis
    plt.grid()
    plt.axis([xmin, xmax, ymin - 2 * dy, ymax + 2 * dy])
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title("Secant Method for Root Finding")
    plt.legend()
    plt.show()

# Example usage
def f(x):
    return np.cos(x) - x / 2

# Call the secant_method function
secant_method(f, x0=0, x1=2.0, xdomain=(-5, 5),niterates= 8)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option("display.precision" ,16)

# xmin= #This is the lower bound of the horizontal axis you should plot
# xmax= #This is the upper bound of the horizontal axis you should plot

#[xmin, xmax] allows as much space as possible for web diagrams based on the iterations produced per each input of
#x0, a, b, and c. Based on this, [xmin, xmax] is an adjustable range to fit the graph. 

plt.figure(figsize=(10,8))
def Myfunction(x0,a,b,n,xmin=-10,xmax=10):  #We define our function of three inputs as Myfunction(x0,a,b,n)
    def f(x):
        return a*x + b
    
    df = pd.DataFrame()  # We put our iterations into a table or a data frame
    xn = x0*np.ones(n+1) #This tells our program to perform n+1 iterations with outputs; x_0, x_1, x_2,...,x_n
    for num in range(n): #Suppose each object in our range of numbers (n) is called num, we perform a for loop to execute each num in n
        xn[num+1] = f(xn[num]) #We're telling our program here that the input for x_k is f(x_k-1)
        
    df['iterates']=xn #Outputs our final iterations for our for loop
    display(df)

    N = range(n+1) #first n iterates
    plt.plot(N,xn,'bo-') #A blue-dotted line plot of N(first n iterates) against outputs
    plt.title('Graphical display of the functional iterates of the function f(x)=kx(1-x)') #Graph title
    plt.xlabel('x-axis')# x-axis label
    plt.ylabel('f(x)')#y-axis label
    plt.grid()# We want to see a grid in our graph
    plt.show()# Now display graph

    #Web diagram

    plt.figure(figsize=(10,8))
    x = np.linspace(xmin,xmax)
    y = f(x)
    plt.plot(x,y,'b--')
    #You need to draw the graph of the line given by y=x for the appropriate region in green and dashed
    plt.plot(x,x,'g--', label='y=x')
    
    # plot initial point (x0,x0) as green circle
    plt.plot(x0,x0, 'go', label='(x0,x0)')
    
    #The for loop below will help us step through the iterates xn of our function and make our plot.
    for num in range(n):
        # plot function evaluation as vert line segment from (xn_j,xn_j) to (xn_j, xn_(j+1))
        plt.plot([xn[num], xn[num]], [xn[num], xn[num+ 1]],'k-')
        
        # plot "output becomes input" as horiz line segment from (xn_j,xn_(j+1)) to (xn_(j+1),xn_(j+1))
        
        plt.plot([xn[num],xn[num + 1]], [xn[num+ 1],xn[num + 1]],'k-')
    
        # plot the (j+1)th iterate (xn_(j+1),xn_(j+1)) as red circle
        plt.plot(xn[num + 1],xn[num + 1],'ro')

    # end of loop 
    plt.axis([xmin,xmax,xmin,xmax])#plots graph within specified limits on the x and y axes
    plt.grid()# displays a grid
    plt.title('Graphical display of the functional iterates of the function f(x)=ax+b') #displays the title of the graph
    plt.legend() #shows the legend of the graph
    plt.xlabel('x-axis') #displays the x-axis of the graph
    plt.ylabel('f(x)')#displays the y-axis of the graph
    # plt.plot(N,xn,'ro')
    plt.show()

Myfunction(2,0.2,0.5,5,-1,3) # Here, we call our function for some inputs x0, a,  b, and n
```
